1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.

```python

# Обчислення похідної
x = sp.symbols('x')
f = 10*x - 5
f_prime = sp.diff(f, x)

print(f"Похідна функції f(x) = 10x - 5: f'(x) = {f_prime}")

# Побудова графіка

x_values = np.linspace(0, 10, 100)
f_values = 10 * x_values - 5
# похідна функції = 10, тому:
f_prime_values = np.full_like(x_values, 10)

plt.figure(figsize=(10,6))
plt.plot(x_values, f_values, label='f(x) = 10x - 5', color='blue')
plt.plot(x_values, f_prime_values, label="f'(x) = 10", color='red')

plt.title('Графік функції та її похідної')
plt.xlabel('Кількість годин навчання (х)')
plt.ylabel('Оцінка (y)')
plt.legend()
plt.legend()
plt.xlim(0, 10)
plt.ylim(-5, 100)
plt.show()
```

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
x = sp.symbols('x')
x_values = np.linspace(0.1, 10, 100)

f_sin = sp.sin(x)
derivative_f_sin = sp.diff(f_sin, x)
print("Похідна Sin(x):", derivative_f_sin)

f_ln = sp.ln(x)
derivative_f_ln = sp.diff(f_ln, x)
print("Похідна ln(x):", derivative_f_ln)

f_sin_values = np.sin(x_values)
f_sin_derivative_values = np.cos(x_values)

f_ln_values = np.log(x_values)
derivative_f_ln_values = 1 / x_values

# Plotting the sine function and its derivative
plt.figure(figsize=(12, 6))

# Sine function plot
plt.subplot(1, 2, 1)
plt.plot(x_values, f_sin_values, label='sin(x)', color='blue')
plt.plot(x_values, f_sin_derivative_values, label="cos(x)", color='red', linestyle='--')
plt.title('Графік функції sin(x) та її похідної cos(x)')
plt.xlabel('x')
plt.ylabel('Значення')
plt.axhline(0, color='black', linewidth=0.5, ls='--')  # Horizontal line at y=0
plt.axvline(0, color='black', linewidth=0.5, ls='--')  # Vertical line at x=0
plt.grid()
plt.legend()

# Natural logarithm plot
plt.subplot(1, 2, 2)
plt.plot(x_values, f_ln_values, label='ln(x)', color='green')
plt.plot(x_values, derivative_f_ln_values, label="1/x", color='orange', linestyle='--')
plt.title('Графік функції ln(x) та її похідної 1/x')
plt.xlabel('x')
plt.ylabel('Значення')
plt.axhline(0, color='black', linewidth=0.5, ls='--')  # Horizontal line at y=0
plt.axvline(0, color='black', linewidth=0.5, ls='--')  # Vertical line at x=0
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

x = sp.symbols('x')

f = x**2 - 5*x + 1

derivative_f = sp.diff(f, x)
print("Похідна функції:", derivative_f)

x_values = np.linspace(0, 100, 100)

f_values = x_values**2 - 5*x_values + 1
f_derivative_values = 2*x_values - 5

# Plotting the function and its derivative
plt.figure(figsize=(10, 6))
plt.plot(x_values, f_values, label='f(x) = x^2 - 5x + 1', color='blue')
plt.plot(x_values, f_derivative_values, label="f'(x) = 2x - 5", color='red')

# Add title and labels
plt.title('Графік функції та її похідної')
plt.xlabel('Кількість вироблених одиниць (x)')
plt.ylabel('Загальні витрати (y)')
plt.legend()
plt.xlim(0, 20)
plt.ylim(-10, 60)
plt.grid()
plt.show()
```
